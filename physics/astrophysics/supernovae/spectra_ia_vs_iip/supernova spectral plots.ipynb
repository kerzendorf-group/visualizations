```python
import numpy as np
```

```python
snia = np.loadtxt('SN2011fe/11feM003.dat')
sniip = np.loadtxt('data/sn1999em_clean.dat')
```

```python
%pylab nbagg
style.use('talk_default')
fig = figure(1)
ax = axes([0.1, 0.15, 0.85, 0.75])
sniip_plot = ax.plot(sniip[:,0]/10., sniip[:,1]*200, color='blue', lw=3, label='Type II')
semilogy()
ax.set_xlim(350, 700)
ax.set_ylim(1e-13, 1e-11)
#ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('Wavelength [nm]')
ax.set_ylabel('Flux')
savefig('sn_spectrum_type2_only.pdf')
snia_plot = ax.plot(snia[:,0]/10., snia[:,1], color='red', lw=3, label='Type Ia')
savefig('sn_spectrum_type2_ia.pdf')
```
